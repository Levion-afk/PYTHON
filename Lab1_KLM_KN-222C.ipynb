```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def get_column_values(file_name, column_name):
    data = pd.read_csv(file_name)
    return np.array(data[column_name])

tank_values = get_column_values('russia_losses_equipment.csv', 'tank')

daily_losses = np.diff(tank_values)

four_largest_losses_indices = np.argsort(daily_losses)[-4:]
four_largest_losses = daily_losses[four_largest_losses_indices]

spring_start_date = '2023-03-01'
spring_end_date = '2023-05-31'
dates = pd.to_datetime(get_column_values('russia_losses_equipment.csv', 'date'))

condition_for_spring_2023 = (dates[:-1] >= spring_start_date) & (dates[:-1] <= spring_end_date)
spring_2023_losses = daily_losses[condition_for_spring_2023]

average_losses_last_100_days = np.mean(daily_losses[-100:])

plt.figure(figsize=(8, 16), dpi=100)
plt.plot(dates[1:], daily_losses, label='Добові втрати', linestyle='dashed', color='blue')
plt.xlabel('Дата')
plt.ylabel('Добові втрати танків')
plt.title('Добові втрати танків за 2023 рік')
plt.legend()
plt.grid(True)
plt.savefig('tank_losses_2023.png')
plt.show()
```
